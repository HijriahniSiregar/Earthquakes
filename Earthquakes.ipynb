# üåç Global Earthquake Data Analysis (1965‚ÄìPresent)

This project explores global earthquake data from 1965 to the present, offering insights into earthquake frequency, magnitude, depth, and geographic distribution.


```python
# üì¶ Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

sns.set(style="whitegrid")

# üìÇ Load dataset
df = pd.read_csv("database.csv")
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Year'] = df['Date'].dt.year
df.head()
```

```python
# üßº Check missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# üìà Magnitude distribution
plt.figure(figsize=(10, 6))
sns.histplot(df["Magnitude"], bins=30, kde=True)
plt.title("Distribution of Earthquake Magnitudes")
plt.xlabel("Magnitude")
plt.ylabel("Frequency")
plt.show()
```

```python
# üìä Number of earthquakes per year
plt.figure(figsize=(12, 6))
df['Year'].value_counts().sort_index().plot(kind='line')
plt.title("Number of Earthquakes Per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
# üåã Depth vs. Magnitude
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Depth', y='Magnitude', alpha=0.5)
plt.title("Depth vs Magnitude")
plt.xlabel("Depth (km)")
plt.ylabel("Magnitude")
plt.show()
```

```python
# üó∫Ô∏è Global earthquake map
fig = px.scatter_geo(df, lat='Latitude', lon='Longitude', color='Magnitude',
                     hover_name="ID", size="Magnitude",
                     projection="natural earth",
                     title="Global Earthquake Map")
fig.show()
```

## üìä Insights

- Most earthquakes are between magnitude 5.0 and 6.0.
- Frequency of recorded earthquakes increased over the decades.
- Shallow earthquakes are more frequent than deep ones.
- Earthquake hotspots are mostly located along tectonic boundaries like the Pacific Ring of Fire.
